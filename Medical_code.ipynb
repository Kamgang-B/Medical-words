```python
import re

def reg(filename):

    with open(filename, "r+") as f,   open("result.txt","w") as q:
        dat=f.readlines()
        wrt=[]
        for i in dat:
            matches = re.findall("[ \w()-]+", i)
            try:
                print(matches[0])
                q.write(f'\n{matches[0]}')
                wrt.append(matches[0])
            except: IndexError
reg("Medi_words.txt")
```
